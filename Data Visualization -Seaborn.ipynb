## Type 8 : Categorital Scatter plot

```python
data.columns
```

```python
# Strip plot
sns.catplot(x='Outlet_Size',y="Item_MRP",data=data[:400],kind="strip")
```

```python
# Strip plot
# transpose by changing x & y
sns.catplot(y='Outlet_Size',x="Item_MRP",data=data[:400],kind="strip")
```

            - Same as strip but data points does not overlap
            - sns.catplot(x,y,data,kind='swarm')

```python
# Swarm plot 
sns.catplot(y='Outlet_Size',x="Item_MRP",data=data[:400],kind="swarm")
```

## Types 10: Density plot     :  <br>
- It uses the KDE method  - Kernel density estimation
- code : sns.kdeplot(data[],shade=True/False)
  - only one col
  - if shade is true then shade comes 
  - if shade false then shade dont appear

            

```python
plt.figure(figsize=(10,5))
sns.kdeplot(data['Item_MRP']) # default shape is false b
```

```python
sns.kdeplot(data['Item_MRP'],shade=True) # default shape is false 
```

### Types 11: Pair plot     :  
#### It gives the scatter plot of every numeric colums with other numeric col 
<br>
- We just have to give rows only as a parameter in code not having x & y 
- It auto generates x & y by considering all numeric cols of dataset
- code : sns.pairplot(data)
  

            

```python
sns.pairplot(data)
```

## Insights from above 
- Just look at 3X3 matrix fro top left 
- scatter plot with all nummeric cols 
- scatter of one col with itself is histogram
- Both sides of the diagonal is the same so look at this trangle or the other

```python
# we can also specify the height parameter in pair plot 
sns.pairplot(data[:300],height=1.0)  # note its in inches
```
